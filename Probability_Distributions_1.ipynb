### Introduction to Probability Distributions

* Probability distributions are a fundamental concept in statistics and probability theory, which play a crucial role in various fields, including mathematics for machine learning (ML). A probability distribution describes how the values of a random variable are distributed. In other words, it provides a mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment.
* Key Concepts
   * Random Variable: A random variable is a variable whose possible values are outcomes of a random phenomenon. There are two types of random variables:

      * Discrete Random Variable: Takes on a countable number of distinct values. Example: The number of heads in 10 coin tosses.
      * Continuous Random Variable: Takes on an infinite number of possible values within a given range. Example: The height of students in a class.
   * Probability Mass Function (PMF): For discrete random variables, the probability distribution is defined by a probability mass function, which gives the probability that a discrete random variable is exactly equal to some value.

   * Probability Density Function (PDF): For continuous random variables, the probability distribution is defined by a probability density function, which describes the likelihood of the random variable taking on a specific value. The probability for a continuous random variable to take a value within a specific interval is given by the area under the curve of the PDF over that interval.

   * Cumulative Distribution Function (CDF): The CDF of a random variable is a function that gives the probability that the variable takes a value less than or equal to a given value.

* Why Probability Distributions are Used in Math for Machine Learning?

   * Probability distributions are fundamental in mathematics for machine learning because they provide a structured way to model uncertainty, which is inherent in real-world data. They enable the estimation of model parameters, support Bayesian inference, and are essential for hypothesis testing, where distributions help evaluate model performance. Moreover, probability distributions are integral to sampling methods and generative models, allowing for the creation of new data points and making predictions with quantified uncertainty. In essence, probability distributions underpin many of the mathematical techniques that allow machine learning models to learn from data, make predictions, and generalize effectively.

#### Geometric Distribution

* The Geometric distribution is a discrete probability distribution that models the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials. Each trial has two possible outcomes: success or failure, with the same probability of success in each trial. The distribution is named for its relationship to geometric probabilities.

* Key Characteristics

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parameters
p = 0.2  # Probability of success

# Create an array of possible trials
k = np.arange(1, 21)

# Compute the PMF of the Geometric distribution
pmf = geom.pmf(k, p)

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.stem(k, pmf, use_line_collection=True)
plt.title('Geometric Distribution PMF')
plt.xlabel('Number of Trials (k)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Compute expected value and variance
expected_value = geom.mean(p)
variance = geom.var(p)

print(f'Expected Value (Mean): {expected_value}')
print(f'Variance: {variance}')
```

#### Negative Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters
r = 5  # Number of successes
p = 0.3  # Probability of success

# Create an array of possible failures
k = np.arange(0, 21)

# Compute the PMF of the Negative Binomial distribution
pmf = nbinom.pmf(k, r, p)

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.stem(k, pmf, use_line_collection=True)
plt.title('Negative Binomial Distribution PMF')
plt.xlabel('Number of Failures (k)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Compute expected value and variance
expected_value = nbinom.mean(r, p)
variance = nbinom.var(r, p)

print(f'Expected Value (Mean): {expected_value}')
print(f'Variance: {variance}')
```

#### Uniform Distribution

* The Uniform distribution is a type of continuous probability distribution where all outcomes within a specified range are equally likely. It is one of the simplest distributions and is used in various applications where each outcome within an interval is equally probable.
* Key Characteristics
1. Constant Probability: In a Uniform distribution, every value within the interval
[a,b] has the same probability density. This means the probability is uniformly spread over the range.

2. Probability Density Function (PDF): The PDF for a continuous Uniform distribution is given by:

   image.png

Outside this range, the PDF is 0.

3. Cumulative Distribution Function (CDF): The CDF, which represents the probability that a random variable

* Example:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a = 0  # Lower bound
b = 1  # Upper bound

# Create an array of possible values within the range [a, b]
x = np.linspace(a - 0.1, b + 0.1, 1000)

# Compute the PDF of the Uniform distribution
pdf = uniform.pdf(x, loc=a, scale=b-a)

# Compute the CDF of the Uniform distribution
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Plot the PDF
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF of Uniform Distribution')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF of Uniform Distribution', color='orange')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Compute expected value and variance
mean = uniform.mean(loc=a, scale=b-a)
variance = uniform.var(loc=a, scale=b-a)

print(f'Expected Value (Mean): {mean}')
print(f'Variance: {variance}')
```

#### Chi-Square Distribution

* The Chi-Square distribution is a continuous probability distribution that is widely used in statistical inference, particularly in hypothesis testing, confidence interval estimation, and goodness-of-fit tests. It is derived from the sum of the squares of

* Example:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters
degrees_of_freedom = 4  # Degrees of freedom

# Create an array of possible values
x = np.linspace(0, 20, 1000)

# Compute the PDF and CDF of the Chi-Square distribution
pdf = chi2.pdf(x, df=degrees_of_freedom)
cdf = chi2.cdf(x, df=degrees_of_freedom)

# Plot the PDF
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, label=f'PDF of Chi-Square Distribution\nk={degrees_of_freedom}')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label=f'CDF of Chi-Square Distribution\nk={degrees_of_freedom}', color='orange')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Compute expected value and variance
mean = chi2.mean(df=degrees_of_freedom)
variance = chi2.var(df=degrees_of_freedom)

print(f'Expected Value (Mean): {mean}')
print(f'Variance: {variance}')
```
