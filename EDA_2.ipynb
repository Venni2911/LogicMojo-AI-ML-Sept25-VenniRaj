# Exploratory Data Analysis Tools

* Exploratory Data Analysis (EDA) is crucial in data science for understanding the characteristics of a dataset before applying any models. Various tools help in performing EDA, ranging from basic statistical techniques to advanced visualizations.
* Here are the commonly used EDA tools:

## Python Libraries

* Pandas:

    * Used for data manipulation and analysis.
    * Functions like .describe(), .info(), .head(), and .value_counts() give quick insights.
    * Built-in functionalities for handling missing data, filtering, grouping, and summarizing data.
* NumPy:

    * Provides support for handling large arrays and matrices of numeric data.
    * Functions like np.mean(), np.median(), and np.std() can calculate basic statistical measures.
* Matplotlib & Seaborn:

    * Matplotlib is a low-level plotting library.
    * Seaborn extends Matplotlib and provides more sophisticated visualization options like histograms, box plots, and violin plots.
    * Both are great for understanding data distributions, trends, and relationships between variables.
* Plotly:

    * Interactive plotting library.
    * Provides advanced visualizations like 3D plots, choropleths, and interactive dashboards.
* SciPy:

    * Focused on scientific computing.
    * Contains modules for statistics, optimization, and integration, used for computing things like correlation, skewness, and statistical tests.

# Use Case for House Price Prediction Dataset

* Objective
   * To predict the sale prices of houses based on various features such as size, location, number of bedrooms, etc.

1. Check for Null Values

   * Purpose: Identify missing values and decide how to handle them (e.g., imputation or removal).
Example: Discover that GarageCars has some missing values.
2. Examine Descriptive Statistics

   * Purpose: Understand the basic statistics (mean, median, standard deviation) of numeric features.
Example: SalePrice has a wide range with a high standard deviation, indicating variability in house prices.
3. Analyze Distributions

   * Purpose: Visualize the distribution of numeric features and the target variable.
Example: Plot a histogram of SalePrice to check for skewness. If skewed, apply a transformation (e.g., log transformation).
4. Check for Correlations

   * Purpose: Identify relationships between features and the target variable.
Example: Create a correlation matrix to see how strongly GrLivArea and OverallQual correlate with SalePrice.
5. Examine Categorical Variables

   * Purpose: Understand the distribution of categorical features and their impact on the target variable.
Example: Analyze the distribution of house prices across different Neighborhood categories using box plots.
6. Detect Outliers

   * Purpose: Identify and decide how to handle outliers in numeric features.
Example: Use scatter plots to find outliers in LotArea or TotalBsmtSF and investigate if they should be removed or transformed.

7. Explore Relationships Between Features

   * Purpose: Understand interactions between features that could affect the target variable.
Example: Use pair plots to examine how combinations of features like OverallQual and GrLivArea influence SalePrice.

### Removing Outliers

```python
data['SalePrice'].describe()
```

```python
sns.boxplot(x=data["SalePrice"])
```

```python
data.shape
```

```python
first_quartile = data['SalePrice'].quantile(0.25)
third_quartile = data['SalePrice'].quantile(0.75)
IQR = third_quartile - first_quartile
```

```python
new_boundarys = third_quartile + 3 * IQR
```

```python
data.drop(data[data['SalePrice'] > new_boundarys].index,axis = 0, inplace=True)
```

Employees recruited from referrals tend to have higher performance scores compared to those recruited from job boards.
Employees from social media platforms like LinkedIn have mid-range performance scores.
       * Next Step: You could investigate if certain recruitment sources lead to higher retention or better job-role fit.
3. Correlation Analysis
   * Objective: Analyze correlations between PerformanceScore and other continuous variables like Tenure, Age, TrainingHours, and Salary.
   * Example:
       * Action: Generate a correlation matrix or heatmap.
       * Insight:

       * In departments like Engineering, employees recruited from referrals perform better.
       * In departments like Sales, performance is relatively uniform across recruitment sources.
   * Next Step: This might suggest that certain recruitment sources work better for specific roles, and the HR team could optimize recruiting efforts accordingly.
5. Outliers Detection
* Objective: Identify outliers in PerformanceScore that could skew the analysis or require further investigation.
* Example:
  * Action: Use a boxplot to identify outliers in PerformanceScore for different JobRoles.
  * Insight:
     * Certain roles like Customer Support have unusually low-performance scores that could be caused by poor job fit, lack of training, or other factors.

*  Examining the Relationship Between Ratings and Box Office Over Time
  * Objective: Analyze whether the relationship between box office performance (BoxOffice) and ratings (e.g., IMDb or Rotten Tomatoes) has changed over time.

     * Action: Create scatter plots of BoxOffice versus IMDbRating, and use color or marker size to represent ReleaseYear. You could also plot the correlation between ratings and box office for different time periods.
