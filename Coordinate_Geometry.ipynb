### Basics of Coordinate Geometry

**What is Coordinate Geometry?**

* Coordinate geometry is a fundamental branch of mathematics that deals with the study of points, lines, curves, and surfaces in a coordinate system. It plays a crucial role in machine learning by providing the mathematical foundation for various algorithms and concepts.

*  Coordinate geometry provides the mathematical framework for representing and manipulating data in machine learning, enabling the development of powerful algorithms and techniques.

   image.png

**Why it is used?**
* Coordinate geometry is used in machine learning for several reasons:
   * Representation: It provides a way to represent data points, vectors, and geometric shapes in a numerical format that can be easily processed by computers.
   * Calculations: It allows for various calculations, such as distances between points, angles between vectors, and finding the equation of lines or curves. These calculations are essential for many machine learning algorithms.
   * Visualization: Coordinate geometry is used to visualize data and the results of machine learning algorithms. This helps in understanding the underlying patterns and relationships within the data.
   * Algorithms: Many machine learning algorithms, such as clustering, support vector machines, and principal component analysis, are based on geometric concepts. These algorithms use coordinate geometry to perform calculations and make decisions.
   * Foundations: Coordinate geometry provides the mathematical foundation for other areas of machine learning, such as linear algebra and vector calculus.


#### Cartesian Coordinates

* Cartesian coordinates are a system for locating points in a plane or space using two or three perpendicular axes.
* These axes are typically labeled as x, y, and z (for 3D space).
* Each point in the plane or space is represented by an ordered pair (x, y) or an ordered triple (x, y, z), respectively.

image.png

* Coordinate Points:
   * Point (x, y): Specifies a location on the plane where
x is the horizontal distance from the origin and
y is the vertical distance from the origin.
   * Example:
The point
(3,4) means you move 3 units to the right of the origin and 4 units up.

* Applications in Machine Learning
   * Data Representation: Cartesian coordinates are used to represent data points in various machine learning algorithms, such as linear regression, clustering, and neural networks.
   * Geometric Transformations: Operations like translation, rotation, and scaling can be represented using matrices and Cartesian coordinates.
   * Distance Calculations: The Euclidean distance between two points can be calculated using their Cartesian coordinates.
   * Visualization: Cartesian coordinates are used to plot data points and visualize relationships between variables.

#####  Origin

* The point of intersection of the x and y axes is called the origin (often denoted by O).
* It represents the point with coordinates (0, 0), where both horizontal and vertical distances are zero.

#### Distance Formula


* The distance formula calculates the straight-line distance between two points
(x<sub>1</sub>, y<sub>1</sub>) and (x<sub>2</sub>, y<sub>2</sub>) in the Cartesian plane.

image.png

* where:

   * d is the distance between the two points
   * (x<sub>1</sub>, y<sub>1</sub>) and (x<sub>2</sub>, y<sub>2</sub>) are the coordinates of the two points.

* Example
   * Let's find the distance between the points (2, 3) and (5, 7) in a 2D plane.

   * Using the distance formula:

                 d = sqrt((5 - 2)^2 + (7 - 3)^2)
                   = sqrt(3^2 + 4^2)
                   = sqrt(9 + 16)
                   = sqrt(25)
                   = 5
So, the distance between the two points is 5 units.

#### Midpoint Formula

* The midpoint formula is used to find the coordinates of the point that is exactly halfway between two given points.

* For two points, (x<sub>1</sub>, y<sub>1</sub>) and (x<sub>2</sub>, y<sub>2</sub>), in a 2D plane, the coordinates of the midpoint are given by:

image.png

* Example
   * Let's find the midpoint of the line segment connecting the points (2, 3) and (5, 7).

  * Using the midpoint formula:

                     (x, y) = ((2 + 5) / 2, (3 + 7) / 2)

#### Slope of a Line

* The slope of a line is a measure of its steepness or inclination. It is defined as the ratio of the vertical change (rise) to the horizontal change (run) between two points on the line.

* The slope is calculated using the following formula:
  * slope = (y<sub>2</sub> - y<sub>1</sub>) / (x<sub>2</sub> - x<sub>1</sub>)
  * Where (x<sub>1</sub>, y<sub>2</sub>) and (x<sub>2</sub>, y<sub>2</sub>) are the coordinates of the two points.
* Interpretation
   * Positive slope: The line slants upward from left to right.
   * Negative slope: The line slants downward from left to right.
   * Zero slope: The line is horizontal.
   * Undefined slope: The line is vertical.
* Example
   * Let's find the slope of the line passing through the points (2, 3) and (5, 7).

   * Using the slope formula:

                         m = (7 - 3) / (5 - 2)
                           = 4 / 3
So, the slope of the line is 4/3. This means that for every 3 units you move horizontally to the right, you move 4 units vertically upward.

### Circle

- A circle is a fundamental shape in geometry, defined as the set of all points in a plane that are equidistant from a fixed point called the center. The constant distance from the center to any point on the circle is called the radius.

image.png

**Key Components of a Circle:**
- **Center**: The fixed point in the plane from which all points on the circle are equidistant.
- Radius (r): The distance from the center to any point on the circle. It remains constant for a given circle.
- **Diameter** (d): The longest distance across the circle, passing through the center. It is twice the radius:
  - d = 2r
- **Circumference** (C): The total distance around the circle. The formula to calculate the circumference is:
  - C=2r
  - where (pi) is approximately 3.14159.
- **Area** (A): The space enclosed by the circle. The formula to calculate the area is:
  - A=r<sup>2</sup>
- **Chord**: A line segment with both endpoints on the circle. A chord that passes through the center is the diameter.
- **Arc**: A portion of the circle's circumference.
  - **Minor Arc:** An arc smaller than a semicircle.
  - **Major Arc:** An arc larger than a semicircle.
- **Sector**: A region enclosed by two radii and the corresponding arc. It looks like a "slice" of the circle.
- **Segment**: A region enclosed by a chord and the arc between the chord's endpoints.

- **Properties of Circles:**
  - Symmetry: A circle is symmetric with respect to its center. Any line through the center is a line of symmetry.
  - Tangent: A line that touches the circle at exactly one point. The tangent is perpendicular to the radius at the point of contact.
  - Secant: A line that intersects the circle at two points.

#### Graphical Representation (Circle)

```python
import matplotlib.pyplot as plt
import numpy as np

# Define circle properties
center_x = 2  # x-coordinate of the center
center_y = 3  # y-coordinate of the center
radius = 5  # radius of the circle

# Generate circle points
theta = np.linspace(0, 2*np.pi, 200)  # Create angles for a smooth circle
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# Create the plot
plt.figure(figsize=(8, 8))

# Plot the circle
plt.plot(x, y, label='Circle')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Circle with Symmetry, Tangent, and Secant')

# Add the center point
plt.scatter(center_x, center_y, color='red', label='Center')

# Symmetry demonstration: Add horizontal and vertical lines through the center
plt.axhline(y=center_y, color='blue', linestyle='--', label='Symmetry Line (Y=Center Y)')
plt.axvline(x=center_x, color='blue', linestyle='--', label='Symmetry Line (X=Center X)')

# Tangent demonstration: Define a point on the circle
tangent_point_x = center_x + radius  # Tangent point on the circle
tangent_point_y = center_y

# Plot the tangent point
plt.scatter(tangent_point_x, tangent_point_y, color='green', label='Tangent Point')

# Add the tangent line: Perpendicular to the radius at the tangent point
plt.plot([tangent_point_x, tangent_point_x], [tangent_point_y - 5, tangent_point_y + 5], color='green', linestyle='-', label='Tangent Line')

# Add the radius to the tangent point to show perpendicularity
plt.plot([center_x, tangent_point_x], [center_y, tangent_point_y], color='orange', linestyle='-', label='Radius')

# Secant demonstration: A line intersecting the circle at two points
secant_x = np.array([center_x - radius, center_x + radius])
secant_y = np.array([center_y - 2, center_y + 2])
plt.plot(secant_x, secant_y, color='purple', linestyle='-', label='Secant Line')

# Add legend
plt.legend()

# Set aspect ratio for a circular shape
plt.axis('equal')

# Show grid and plot
plt.grid(True)
plt.show()
```

### Hyperplane

* A hyperplane is a subspace whose dimensionality is one less than its surrounding space. In simpler terms:

   * In a 2-dimensional (2D) space (like a flat piece of paper), a hyperplane is a line (1D).
   * In a 3-dimensional (3D) space (like the physical world we experience), a hyperplane is a plane (2D).
   * In a n-dimensional space, a hyperplane is an (n1)-dimensional subspace. So in a 4D space, a hyperplane would be a 3D space.

* Equation of a Hyperplane
The equation of a hyperplane in an
n-dimensional space can be written as:

        w1x1 + w2x2 + +wnxn + b = 0
Where:
   * w1, w2,,wn  are the weights or coefficients (which define the orientation of the hyperplane).
   * x1, x2,, xn  are the features or coordinates in n-dimensional space.
   * b is the bias term or intercept (which controls the offset of the hyperplane from the origin).
* In 2D (for two features,
x
1 and x2), this reduces to the familiar equation of a line:

          w1x1 + w2x2 + b = 0
* In 3D (for three features,
x1, x2, x3), this becomes the equation of a plane:

         w1x1 + w2x2 + w3x3 + b = 0

Example: SVM with Hyperplane Visualization
* In this example, we'll:

   * Generate a sample dataset with two features.
   * Train a Support Vector Machine (SVM) classifier.
   * Visualize the decision boundary (hyperplane) and the support vectors.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Step 1: Generate a simple dataset with two classes
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6, cluster_std=1.5)

# Step 2: Train the SVM model
# We'll use a linear kernel, as it gives a linear hyperplane
model = SVC(kernel='linear')
model.fit(X, y)

# Step 3: Plot the data points and the hyperplane
# Create a function to plot the decision boundary and margins
def plot_hyperplane(clf, X, y):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)

    # Get the separating hyperplane
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Create a grid to plot the decision boundary
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)

    # Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.7, linestyles=['--', '-', '--'])

    # Plot support vectors
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')

    plt.title('SVM: Hyperplane and Support Vectors')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Step 4: Call the function to plot the hyperplane
plot_hyperplane(model, X, y)
```

### Parabola

* Properties of Parabolas

   * Directrix: A fixed line outside the parabola.
   * Axis of symmetry: A line that divides the parabola into two symmetrical halves.
   * Vertex: The point where the parabola intersects its axis of symmetry.

#### Graphical Representation of Parabola

```python
import matplotlib.pyplot as plt

# Define parabola properties (for standard upward facing parabola)
vertex_x = 2  # x-coordinate of the vertex
vertex_y = 1  # y-coordinate of the vertex
focus_offset = 0.5  # Distance between the focus and the vertex (controls shape)

# Calculate focus coordinates (assuming upward facing parabola)
focus_x = vertex_x
focus_y = vertex_y + focus_offset  # Focus is above the vertex for upward parabola

# Function to generate parabola points
def generate_parabola_points(vertex_x, vertex_y, focus_y, num_points):
  """Generates coordinates for points on the parabola.

  Args:
      vertex_x (float): x-coordinate of the vertex.
      vertex_y (float): y-coordinate of the vertex.
      focus_y (float): y-coordinate of the focus.
      num_points (int): Number of points to generate.

  Returns:
      list: List of tuples containing (x, y) coordinates of points on the parabola.
  """
  points = []
  a = 1 / (2 * focus_offset)  # Parameter based on focus and vertex position
  for i in range(-num_points, num_points + 1):
    x = vertex_x + i
    y = vertex_y + a * (x - vertex_x)**2
    points.append((x, y))
  return points

# Generate points
points = generate_parabola_points(vertex_x, vertex_y, focus_y, 50)

# Extract x and y coordinates for plotting
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]

# Plot the parabola
plt.plot(x_values, y_values, label='Parabola')

# Plot additional elements (optional)
plt.scatter(vertex_x, vertex_y, color='red', label='Vertex')
plt.scatter(focus_x, focus_y, color='green', label='Focus')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Parabola')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Ellipse

* Definition:
   * An ellipse is the set of all points in a plane where the sum of the distances from two fixed points (called foci) is constant.
   * Ellipses are oval-shaped curves and can be thought of as a "stretched" circle.

image.png

* Basic Properties:
   * Center: The midpoint of the line segment joining the two foci.
   * Foci (singular: focus): Two fixed points on the interior of the ellipse such that the sum of the distances to any point on the ellipse is constant.
   * Major Axis: The longest diameter of the ellipse, passing through both foci.
   * Minor Axis: The shortest diameter of the ellipse, perpendicular to the major axis.
   * Vertices: The points where the ellipse intersects the major axis.
   * Co-vertices: The points where the ellipse intersects the minor axis.
  
* Standard Equation of an Ellipse
   * The standard form of the ellipse equation depends on the orientation of the ellipse (horizontal or vertical).

  * Ellipse with Horizontal Major Axis:
     * Equation:

      image.png

* where:
      
   * (h,k) is the center of the ellipse.
   * a is the semi-major axis (half the length of the major axis).
   * b is the semi-minor axis (half the length of the minor axis).
   * If a>b, the major axis is horizontal.

* Example: For an ellipse centered at
(0,0) with
a=5 and
b=3, the equation is:

      (x^2/25)+(y^2/9)=1



#### Graphical Representation of Ellipse

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define ellipse properties
center_x = 2  # x-coordinate of the center
center_y = 3  # y-coordinate of the center
major_radius = 5  # Radius along the major axis
minor_radius = 3  # Radius along the minor axis

# Function to generate ellipse points
def generate_ellipse_points(center_x, center_y, major_radius, minor_radius, num_points):
  """Generates coordinates for points on the ellipse.

  Args:
      center_x (float): x-coordinate of the center.
      center_y (float): y-coordinate of the center.
      major_radius (float): Radius along the major axis.
      minor_radius (float): Radius along the minor axis.
      num_points (int): Number of points to generate.

  Returns:
      list: List of tuples containing (x, y) coordinates of points on the ellipse.
  """
  points = []
  theta = np.linspace(0, 2*math.pi, num_points)  # Generate angles for even distribution
  for angle in theta:
    x = center_x + major_radius * math.cos(angle)
    y = center_y + minor_radius * math.sin(angle)
    points.append((x, y))
  return points

# Generate points
num_points = 100
points = generate_ellipse_points(center_x, center_y, major_radius, minor_radius, num_points)

# Extract x and y coordinates for plotting
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]

# Plot the ellipse
plt.plot(x_values, y_values, label='Ellipse')

# Plot the center (optional)
plt.scatter(center_x, center_y, color='red', marker='o', label='Center')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Ellipse')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.axis('equal')  # Ensure equal aspect ratio for circular representation
plt.show()
```

### Hyperbola

* A hyperbola is the set of all points in a plane where the absolute difference of the distances from two fixed points (called foci) is constant.

image.png

* Key Properties:
   * Center: The midpoint of the line segment joining the two foci.
   * Foci (singular: focus): Two fixed points used to define the hyperbola.The distance from any point on the hyperbola to each focus differs by a constant amount.
   * Vertices: The points where the hyperbola intersects its transverse axis.
   * Transverse Axis: The line segment joining the two vertices. It lies along the major axis of the hyperbola.
   * Conjugate Axis: A line perpendicular to the transverse axis that helps define the box enclosing the asymptotes of the hyperbola.
   * Asymptotes: Straight lines that the branches of the hyperbola approach but never touch. These lines help guide the shape of the hyperbola.
  
* Standard Equations of a Hyperbola:

1. Horizontal Hyperbola:
  * Equation:
    
    image.png


* The transverse axis is along the x-axis (horizontal direction).
* a is the distance from the center to each vertex.
* b is the distance from the center to a point on the conjugate axis.
* (h,k) is the center of the hyperbola.

2. Vertical Hyperbola:
  * Equation:

  image.png

* The transverse axis is along the y-axis (vertical direction).
* a,b, and (h,k) have the same meanings as in the horizontal hyperbola case.

#### Graphical Representation of Hyperbola

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the hyperbola
a = 2  # distance from center to vertex along x-axis
b = 1  # distance from center to vertex along y-axis

# Generate values for x
x = np.linspace(-5*a, 5*a, 400)

# Calculate corresponding y values for the upper branch of the hyperbola
y_upper = np.sqrt((x**2 / a**2 - 1) * b**2)

# Calculate corresponding y values for the lower branch of the hyperbola
y_lower = -np.sqrt((x**2 / a**2 - 1) * b**2)

# Plot the hyperbola
plt.figure(figsize=(8, 6))
plt.plot(x, y_upper, 'r', label='Upper Branch')
plt.plot(x, y_lower, 'b', label='Lower Branch')
plt.plot(x, -y_upper, 'r')
plt.plot(x, -y_lower, 'b')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Standard Hyperbola')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.axis('equal')
plt.show()
```

#### Using Hyperplane

* Using hyperplane to check whether it is passing from the origin or not.
* If hyperplane is passing from the origin the w0 = 0.
* Then,
      W^T.x = 0 --> Passing through Origin

Lets use Support Vector Machines (SVM) to classify the toys based on their length and width. We will create a hyperplane that separates hard and soft toys.

```python
# Create the figure
fig = plt.figure()

# add axes
ax = fig.add_subplot(111,projection='3d')

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

xx, yy = np.meshgrid(x,y)

z = (4/2)*xx - (5/2)*yy

#plot the plane
ax.plot_surface(xx,yy,z,alpha=0.5)

#plot origin
ax.scatter(0,0,0, color='green')

plt.show()
```

At one point of the plane all the data points will be 0,0,0.

#### Distance Between Two Toys

You can calculate the distance between two toy points in 2D (based on length and width) using the Euclidean distance formula:

image.png

```python
import matplotlib.pyplot as plt
from math import sqrt

# Points representing length and width of two toys
toy1 = [0.94414894, 1.13508341]  # Toy 1
toy2 = [0.71809875, 0.5969006]    # Toy 2

# Calculate Euclidean distance
distance = sqrt((toy2[0] - toy1[0])**2 + (toy2[1] - toy1[1])**2)
print(f"Distance between toy1 and toy2: {distance:.2f}")

# Create a plot to visualize the points and the distance
plt.figure(figsize=(6, 6))
plt.scatter(toy1[0], toy1[1], color='blue', label=f"Toy 1 ({toy1[0]:.2f}, {toy1[1]:.2f})", s=100)
plt.scatter(toy2[0], toy2[1], color='red', label=f"Toy 2 ({toy2[0]:.2f}, {toy2[1]:.2f})", s=100)

# Plot the line connecting the two toys to show the distance
plt.plot([toy1[0], toy2[0]], [toy1[1], toy2[1]], color='green', linestyle='--', label=f"Distance = {distance:.2f}")

# Add labels and title
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Euclidean Distance Between Two Toys')
plt.grid(True)

# Annotate the points
plt.text(toy1[0], toy1[1], '  Toy 1', verticalalignment='bottom', horizontalalignment='right')
plt.text(toy2[0], toy2[1], '  Toy 2', verticalalignment='bottom', horizontalalignment='left')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

####  Finding the Midpoint Between Two Toys

The midpoint between two toys in terms of length and width can be calculated using the formula:

image.png

```python
import matplotlib.pyplot as plt

# Toy coordinates (length, width)
toy1 = [0.94414894, 1.13508341]  # Toy 1
toy2 = [0.71809875, 0.5969006]  # Toy 2

# Calculate the midpoint between toy1 and toy2
midpoint = [(toy1[0] + toy2[0]) / 2, (toy1[1] + toy2[1]) / 2]
print(f"Midpoint between toy1 and toy2: {midpoint}")

# Create a plot to visualize the points, the line, and the midpoint
plt.figure(figsize=(6, 6))
plt.scatter(toy1[0], toy1[1], color='blue', label="Toy 1 (0.94414894, 1.13508341)", s=100)
plt.scatter(toy2[0], toy2[1], color='red', label="Toy 2 (0.71809875, 0.5969006)", s=100)

# Plot the line connecting the two toys
plt.plot([toy1[0], toy2[0]], [toy1[1], toy2[1]], color='green', label="Line Between Toys")

# Plot the midpoint
plt.scatter(midpoint[0], midpoint[1], color='purple', marker='x', s=100, label=f"Midpoint ({midpoint[0]:.1f}, {midpoint[1]:.1f})")

# Add labels and title
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('Visualizing the Midpoint Between Two Toys')
plt.grid(True)

# Annotate the points and midpoint
plt.text(toy1[0], toy1[1], '  Toy 1', verticalalignment='bottom', horizontalalignment='right')
plt.text(toy2[0], toy2[1], '  Toy 2', verticalalignment='bottom', horizontalalignment='left')
plt.text(midpoint[0], midpoint[1], '  Midpoint', verticalalignment='bottom', horizontalalignment='left')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#### Slope of a Line Between Two Toys

The slope of the line connecting two toys based on their length and width can be computed as:

image.png

```python
import matplotlib.pyplot as plt

# Toy coordinates (length, width)
toy1 = [0.94414894, 1.13508341]  # Toy 1
toy2 = [0.71809875, 0.5969006]  # Toy 2

# Calculate the slope between toy1 and toy2
slope = (toy2[1] - toy1[1]) / (toy2[0] - toy1[0])
print(f"Slope between toy1 and toy2: {slope:.2f}")

# Create a plot to visualize the points and the slope line
plt.figure(figsize=(6, 6))
plt.scatter(toy1[0], toy1[1], color='blue', label="Toy 1 (0.94414894, 1.13508341)", s=100)
plt.scatter(toy2[0], toy2[1], color='red', label="Toy 2 (0.71809875, 0.5969006)", s=100)

# Plot the line connecting the two toys (representing the slope)
plt.plot([toy1[0], toy2[0]], [toy1[1], toy2[1]], color='green', label=f"Slope = {slope:.2f}")

# Add labels and title
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('Visualizing the Slope Between Two Toys')
plt.grid(True)

# Annotate the points
plt.text(toy1[0], toy1[1], '  Toy 1', verticalalignment='bottom', horizontalalignment='right')
plt.text(toy2[0], toy2[1], '  Toy 2', verticalalignment='bottom', horizontalalignment='left')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
