* Example: Simulating a Random Experiment of Rolling a Die
Let's simulate a random experiment of rolling a six-sided die.
   *  Each roll is a random experiment where the outcome can be any of the numbers 1 through 6, and each outcome is equally likely.

```python
import random

# Define the sample space for the die
sample_space = [1, 2, 3, 4, 5, 6]

# Define the number of times you want to roll the die
num_experiments = 10

# Simulate the random experiment
outcomes = []
for _ in range(num_experiments):
    roll = random.choice(sample_space)  # Randomly choose an outcome from the sample space
    outcomes.append(roll)

# Display the results
print(f"Results of {num_experiments} die rolls: {outcomes}")

# Calculate the frequency of each outcome
frequency = {outcome: outcomes.count(outcome) for outcome in sample_space}
print("Frequency of each outcome:")
for outcome, count in frequency.items():
    print(f"{outcome}: {count}")

# Calculate the empirical probability of each outcome
probability = {outcome: count / num_experiments for outcome, count in frequency.items()}
print("Empirical probability of each outcome:")
for outcome, prob in probability.items():
    print(f"{outcome}: {prob:.2f}")
```

#### Event

* Definition: An event is a subset of the sample space. It consists of one or more outcomes.
* In the context of machine learning, an event is a specific outcome or set of outcomes from an experiment or observation. It's a fundamental concept in probability theory, which forms the basis of many machine learning algorithms.
* Types of Events:
   * Simple Event: An event that consists of a single outcome. For example, in a coin toss, getting a "heads" is a simple event.
   * Compound Event: An event that consists of multiple outcomes. For instance, in a card game, drawing a face card (jack, queen, or king) is a compound event.
   * Mutually Exclusive Events: Events that cannot occur simultaneously. For example, flipping a coin and getting both "heads" and "tails" at the same time is impossible.
   * Exhaustive Events: A set of events that covers all possible outcomes of an experiment. In a coin toss, the events "heads" and "tails" are exhaustive.
* Example: Rolling a Die and Checking for an Event (Rolling an Even Number)
   * Let's consider the random experiment of rolling a six-sided die. We will define an event
ùê¥
A as rolling an even number.

```python
import random

# Define the sample space for the die
sample_space = [1, 2, 3, 4, 5, 6]

# Define event A: rolling an even number
event_A = [2, 4, 6]

# Simulate rolling the die
die_roll = random.choice(sample_space)
print(f"Rolled: {die_roll}")

# Check if the roll is part of event A (rolling an even number)
if die_roll in event_A:
    print("Event A occurred: Rolled an even number.")
else:
    print("Event A did not occur: Rolled an odd number.")
```

#### Dependent Events

* Two events are dependent if the occurrence of one event affects the probability of the other.
* Example: Drawing Cards from a Deck Without Replacement
   * Consider drawing two cards from a standard deck of 52 playing cards without replacement. The probability of drawing a specific card on the second draw depends on the outcome of the first draw, making these events dependent.

```python
import random

# Create a standard deck of 52 cards (simplified as numbers 1 to 52)
deck = list(range(1, 53))

# Shuffle the deck to randomize it
random.shuffle(deck)

# Draw the first card
first_card = deck.pop()  # Removes and returns the last card from the deck
print(f"First card drawn: {first_card}")

# Draw the second card
second_card = deck.pop()  # Removes and returns the last card from the remaining deck
print(f"Second card drawn: {second_card}")

# The deck now has 50 cards
print(f"Remaining cards in the deck: {len(deck)}")
```

### Bayes' Theorem

* Bayes' Theorem is a fundamental principle in probability theory that provides a way to update probabilities as new evidence becomes available. It's a cornerstone of many machine learning algorithms, particularly those based on probabilistic models.

image.png

* Where:

   * P(A|B): The probability of event A occurring given that event B has occurred.
   * P(B|A): The probability of event B occurring given that event A has occurred.
   * P(A): The prior probability of event A.
   * P(B): The prior probability of event B.
* Example: Medical Diagnosis
   * Suppose a certain disease affects 1% of a population. A test for the disease is 90% accurate, meaning it correctly identifies 90% of those with the disease (true positive) and correctly identifies 90% of those without the disease (true negative). Given that a person tests positive, we want to calculate the probability that they actually have the disease.

```python
# Given probabilities
P_disease = 0.01  # Prior probability of having the disease (1% of the population)
P_no_disease = 1 - P_disease  # Probability of not having the disease

P_positive_given_disease = 0.90  # Probability of testing positive given that the person has the disease (True Positive Rate)
P_positive_given_no_disease = 0.10  # Probability of testing positive given that the person does not have the disease (False Positive Rate)

# Calculate the total probability of testing positive
P_positive = (P_positive_given_disease * P_disease) + (P_positive_given_no_disease * P_no_disease)

# Apply Bayes' Theorem to calculate the probability of having the disease given a positive test result
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

# Display the result
print(f"Probability of having the disease given a positive test result: {P_disease_given_positive:.4f}")
```

### Random Variable


* A random variable is a mathematical entity that represents the outcome of an experiment. In the context of machine learning, it's often used to model uncertain quantities or variables that can take on different values.
* Types of Random Variables:
   



#### 1. Discrete Random Variable:
* Takes on a countable number of values (e.g., number of heads in 3 coin flips).
* Example: Rolling a Die
   * Let's consider the example of rolling a fair six-sided die. The discrete random variable
ùëã
X represents the outcome of the roll, which can be any of the integers from 1 to 6. We will simulate rolling the die multiple times, calculate the probability distribution, and then compute the expected value.

```python
import random
from collections import Counter

# Define the sample space for the die
sample_space = [1, 2, 3, 4, 5, 6]

# Simulate rolling the die 1000 times
num_rolls = 1000
rolls = [random.choice(sample_space) for _ in range(num_rolls)]

# Count the frequency of each outcome
frequency = Counter(rolls)

# Calculate the probability distribution
probability_distribution = {outcome: frequency[outcome] / num_rolls for outcome in sample_space}

# Display the probability distribution
print("Probability Distribution:")
for outcome, probability in probability_distribution.items():
    print(f"P(X={outcome}) = {probability:.3f}")

# Calculate the expected value (mean) of the discrete random variable
expected_value = sum(outcome * probability for outcome, probability in probability_distribution.items())
print(f"\nExpected Value (E[X]): {expected_value:.3f}")
```

* Takes on an infinite number of values (e.g., the exact height of students in a class).
* Example: Normal Distribution

   * We'll simulate random values from a normal distribution, plot the probability density function, and compute probabilities for specific ranges.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate random samples from the normal distribution
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Plot the histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot the probability density function (PDF)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', lw=2)

# Display the plot
plt.title('Normal Distribution (Mean = 0, Std Dev = 1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()

# Calculate the probability of X being within a specific range (e.g., -1 to 1)
lower_bound = -1
upper_bound = 1
probability = norm.cdf(upper_bound, mu, sigma) - norm.cdf(lower_bound, mu, sigma)
print(f"Probability that X is between {lower_bound} and {upper_bound}: {probability:.4f}")
```

### Expectation and Variance

#### Expectation (Mean)

* Expectation is a measure of the central tendency of a random variable. It represents the average value that a random variable is expected to take over many trials.

image.png

* In machine learning:

   * Mean of a dataset: The expectation of a random variable representing the data points.
   * Expected value of a prediction: The average value a model is expected to predict for a given input.

* Example: Rolling a Fair Die
    * Let's calculate the expectation (mean) of a discrete random variable representing the outcome of rolling a fair six-sided die.   

```python
import numpy as np

# Define the possible outcomes of rolling a die
outcomes = np.array([1, 2, 3, 4, 5, 6])

# Define the probability of each outcome (for a fair die, each outcome has the same probability)
probabilities = np.array([1/6] * 6)

# Calculate the expectation (mean) of the discrete random variable
mean = np.sum(outcomes * probabilities)
print(f"Expected Value (Mean) of rolling a fair die: {mean:.2f}")
```

#### Variance

* Variance is a measure of how spread out the values of a random variable are. It quantifies the dispersion of the data around the mean.
image.png

* In machine learning:

   * Standard deviation of a dataset: The square root of the variance, which gives a measure of dispersion in the same units as the data.
   * Error variance: A measure of how much a model's predictions deviate from the true values.

* Variance Calculation
   * Let's calculate the variance of a small dataset.   

```python
import numpy as np

# Define the dataset
data = np.array([2, 4, 6, 8, 10])

# Calculate the sample mean
mean = np.mean(data)

# Calculate the variance
variance = np.var(data, ddof=1)  # ddof=1 for sample variance

print(f"Mean of the dataset: {mean:.2f}")
print(f"Variance of the dataset: {variance:.2f}")
```

```python
import numpy as np

# Define the two random variables
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])

# Calculate the means of X and Y
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Calculate the covariance
covariance = np.mean((X - mean_X) * (Y - mean_Y))

print(f"Covariance between X and Y: {covariance:.2f}")
```
