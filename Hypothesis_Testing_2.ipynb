### A/B testing

* A/B testing, also known as split testing, is a method used to compare two versions of a web page, app, or any digital product to determine which one performs better based on a specific metric, such as click-through rate, conversion rate, or user engagement. It's widely used in marketing, UX design, and product development to make data-driven decisions and optimize user experience.

   image.png

* Key Concepts in A/B Testing:
   * Hypothesis Formation:

      * Identify a problem or opportunity: Understand what you want to improve (e.g., "We believe changing the call-to-action button color will increase conversions").
      * Formulate a hypothesis: Create a clear and testable hypothesis (e.g., "If we change the button color from green to red, then the conversion rate will increase").
   * Control and Variation:

      * Control (A): This is the original version of the element youre testing (e.g., the current version of a web page).
      * Variation (B): This is the modified version youre testing against the control (e.g., the web page with the red button).
   * Randomization:

      * Users are randomly assigned to either the control or the variation group to ensure that any differences in performance can be attributed to the changes made, not to other external factors.
   * Sample Size:

      * A sufficient number of users need to participate in the test to ensure statistical validity. The sample size should be large enough to detect a meaningful difference between the control and variation.
   * Duration:

      * The test should run for a period long enough to collect sufficient data but not so long that external factors (like seasonal trends) might influence the results.
   * Metrics:

      * Primary Metric: The main goal youre measuring (e.g., conversion rate, click-through rate).
      * Secondary Metrics: Other related metrics that might be affected by the test (e.g., bounce rate, time on page).
   * Statistical Significance:

      * After the test, statistical analysis is used to determine whether the observed differences in performance are likely to be real or just due to random chance. Common thresholds for significance are p-values of 0.05 or lower.

   * Analysis:

      * Determine the winner: Compare the performance of the control and variation using your chosen metrics. If the variation outperforms the control significantly, its considered the winner.
      * Learn from the results: Even if the variation doesnt outperform the control, theres valuable insight to be gained from understanding why.
   * Implementation:

      * If the variation is successful, it is typically implemented as the new control. The process can be repeated with further optimizations.

* Types of A/B Testing:
  * A/B/n Testing:

      * Involves more than one variation (e.g., A/B/C testing where there are two variations against the control).

  * Multivariate Testing (MVT):

      * Tests multiple elements on a page simultaneously to understand the interaction between them (e.g., testing different combinations of headlines, images, and button colors).

  * Split URL Testing:

      * Redirects a portion of your traffic to a different URL where a completely different page or design is tested against the original.

* Example of an A/B Test:

Scenario: An e-commerce website wants to increase its newsletter sign-ups.

  * Hypothesis: Changing the sign-up button text from "Subscribe" to "Join our community" will increase the sign-up rate.
  * Control (A): The current sign-up button text is "Subscribe".
  * Variation (B): The sign-up button text is changed to "Join our community".
  * Metric: The primary metric is the percentage of visitors who sign up for the newsletter.
  * Randomization: Visitors are randomly shown either the control or the variation.
  * Duration: The test runs for two weeks.
  * Analysis: After two weeks, the variation shows a statistically significant 10% increase in sign-up rates compared to the control.
  * Conclusion: The new button text is adopted, and the test results are used to inform future A/B tests.

### Chi-Square Testing

* Chi-square testing is a non-parametric statistical test used to determine if there is a significant association between categorical variables or if a sample distribution fits an expected distribution. There are two main types of chi-square tests: the Chi-Square Test of Independence and the Chi-Square Goodness-of-Fit Test.

* Steps Involved in a Chi-Square Test:
1. State the null and alternative hypotheses:
   * Null hypothesis (H0): There is no significant difference between observed and expected frequencies.
   * Alternative hypothesis (H1): There is a significant difference between observed and expected frequencies.
2. Set the significance level (): This is typically set to 0.05, but it can be adjusted based on the desired level of confidence.
3. Calculate the chi-square statistic:
   * For each category, calculate the expected frequency using the formula: Expected frequency = (row total * column total) / grand total.
   * Calculate the chi-square statistic using the formula:  =  [(O - E) / E], where O is the observed frequency and E is the expected frequency.
4. Determine the degrees of freedom: For a chi-square test of independence, the degrees of freedom are calculated as (rows - 1) * (columns - 1). For a chi-square test of goodness of fit, the degrees of freedom are calculated as the number of categories minus 1.
5. Find the critical value: Use a chi-square distribution table to find the critical value based on the degrees of freedom and the chosen significance level.
6. Compare the calculated chi-square statistic to the critical value: If the calculated chi-square statistic is greater than the critical value, reject the null hypothesis. Otherwise, fail to reject the null hypothesis.

* Types of Chi-Square Tests

There are two main types of chi-square tests:

   * Chi-Square Test of Independence: This test is used to determine if there is a relationship between two categorical variables. For example, you could use this test to determine if there is a relationship between gender and voting preference.
   * Chi-Square Test of Goodness of Fit: This test is used to determine if a set of observed frequencies fits a particular distribution. For example, you could use this test to determine if the distribution of coin flips is uniform.

#### Chi-Square Test of Independence

* The Chi-Square Test of Independence is a statistical method used to determine if there is a significant relationship between two categorical variables. It helps to assess whether the observed frequencies of the categories are independent or if there is a dependency between them.

* When to Use:
   * Categorical Data: Both variables must be categorical (e.g., gender, occupation, color, etc.).
   * Independence Assessment: To determine if there is a relationship or association between the two variables.

* Steps Involved:
1. Create a Contingency Table: Organize the data into a two-way table, where the rows represent the categories of one variable and the columns represent the categories of the other variable.
2. Calculate Expected Frequencies: For each cell in the contingency table, calculate the expected frequency using the formula:
Expected Frequency = (Row Total * Column Total) / Grand Total
3. Calculate the Chi-Square Statistic: Use the formula:
 =  [(Observed Frequency - Expected Frequency) / Expected Frequency]
Sum over all cells in the contingency table.

* Why is Power Analysis Important?
   * Determining Sample Size: Power analysis helps researchers determine the minimum sample size required to detect an effect of a given size with a certain level of confidence. Insufficient power can lead to inconclusive results.
   * Avoiding Type II Errors: Adequate power reduces the probability of failing to detect a true effect, thus minimizing Type II errors.
   * Study Planning: Power analysis is used in the design phase of a study to ensure that the study is well-equipped to answer the research question.

* Limitations of Power Analysis
   * Assumptions: Power analysis requires assumptions about the effect size and variance, which may not be accurate.

* Conducting Power Analysis
   * Power analysis can be conducted either a priori (before data collection) to determine the necessary sample size or post hoc (after data collection) to evaluate the power of the study based on the collected data.

1. A Priori Power Analysis
  * Performed before collecting data to estimate the minimum sample size needed.
  * Requires assumptions about effect size, significance level, and desired power.

Example (A Priori Power Analysis for a t-test):

```python
import statsmodels.stats.power as smp

# Define parameters
effect_size = 0.5   # Medium effect size (Cohen's d)
alpha = 0.05        # Significance level
power = 0.8         # Desired power (80%)

# Calculate required sample size
analysis = smp.TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power)

print(f"Required sample size: {sample_size}")
```

2. Post Hoc Power Analysis
   * Conducted after the study to determine the achieved power.
   * Helps in assessing whether non-significant results might be due to insufficient power.
   
Example (Post Hoc Power Analysis for a t-test):

```python
# Assuming the study had 50 subjects per group
sample_size_per_group = 50

# Calculate the achieved power
achieved_power = analysis.solve_power(effect_size=effect_size, alpha=alpha, nobs1=sample_size_per_group)

print(f"Achieved power: {achieved_power}")
```

* Calculation of Power of a test

```python
# calculation of power of a test
# Importing the TTestPower module from statsmodels.stats.power, which provides tools for power analysis and sample size calculation specific to t-tests.
from statsmodels.stats.power import TTestPower

# Creating an instance of the TTestPower class from statsmodels.stats.power, enabling further analysis and calculations for power and sample size in the context of t-tests.
power = TTestPower()


# Using the solve_power method from the TTestPower instance to calculate the required sample size for a t-test.
# Parameters: initial sample size (nobs), effect size (0.8), desired power (None, as it is being calculated), and significance level (alpha=0.05).
n_test = power.solve_power(nobs=24, effect_size=0.8, power=None, alpha=0.05)

print('Power: {:.3f}'.format(n_test))
```

**Interpreting Power Analysis Results**

* High Power (e.g., > 0.8): Indicates a high likelihood of detecting an effect if it exists. Large sample sizes and/or large effect sizes contribute to high power.
* Low Power (e.g., < 0.5): Suggests a high risk of Type II error, meaning the study may fail to detect an effect even if one exists. This could be due to a small sample size or a small effect size.
* Balanced Power: Researchers aim for a balance between Type I and Type II errors, typically aiming for a power of 0.8, which is considered a good trade-off.

#### Multiple Comparisons (Multiple Testing)

* Multiple comparisons (or multiple testing) refer to situations in statistical analysis where multiple hypotheses are tested simultaneously. This is common in fields like genomics, psychology, and data science, where numerous statistical tests are conducted within a single study. However, performing multiple tests increases the risk of encountering Type I errors (false positives).

* The Problem of Multiple Comparisons
   * When conducting multiple tests, the probability of making a Type I error (falsely rejecting a true null hypothesis) increases. This is because each test has a certain probability of making a Type I error. As the number of tests increases, so does the overall probability of making at least one Type I error.

* Consequences of Multiple Comparisons
   * False Positives: An increased likelihood of finding significant results that are not actually real.
   * Inflated Type I Error Rate: The overall Type I error rate becomes higher than the nominal level (e.g., 0.05) set for individual tests.
   * Misleading Conclusions: False positive results can lead to incorrect conclusions and wasted resources.

* Methods to Control Multiple Comparisons

Several methods have been developed to control the overall Type I error rate when conducting multiple comparisons:

   * Bonferroni Correction: This is a simple method that divides the desired overall significance level (e.g., 0.05) by the number of tests. Each individual test must have a p-value less than the adjusted significance level to be considered significant.
   * False Discovery Rate (FDR): FDR controls the proportion of false positives among all rejected null hypotheses. It's a more flexible approach than Bonferroni correction and can be more powerful in some cases.
   * Holm-Bonferroni Correction: This is a step-down procedure that adjusts the significance level for each test based on the rank of its p-value.
   * Tukey's HSD (Honestly Significant Difference): This method is specifically designed for comparing multiple means and controls the family-wise error rate.
   * Sidak Correction:
      * The Sidak correction is an alternative to the Bonferroni correction that is slightly less conservative.

#### Effect size

Unlike p-values, which tell you whether an effect exists, effect sizes tell you how large or small that effect is. Understanding effect size is crucial because it provides context to the significance of the findings, enabling researchers to assess the practical or real-world importance of their results.

 image.png

   * Magnitude of Effect: It provides a scale of the magnitude of the effect, helping to determine if the effect is trivial, small, moderate, or large.

   * Comparability: Effect sizes allow for the comparison of results across different studies, even if the sample sizes or scales are different.

   * Practical Significance: While p-values indicate statistical significance, effect sizes indicate practical significance, showing whether the effect is meaningful in real-world terms.

#### Non-Parametric Tests

* Non-parametric tests are statistical tests that do not assume a specific distribution for the data. Unlike parametric tests, which require assumptions such as normality, non-parametric tests are more flexible and can be applied to data that do not meet these assumptions. They are particularly useful when dealing with small sample sizes, ordinal data, or data that are not normally distributed.

   image.png

* Why Use Non-Parametric Tests?
   * No Assumption of Normality: Non-parametric tests do not require the assumption that the data follow a normal distribution.

   * Applicable to Ordinal Data: These tests can be used for data measured on an ordinal scale, where the data are ranked rather than measured on an interval or ratio scale.

   * Robust to Outliers: Non-parametric tests are less sensitive to outliers, making them robust in the presence of non-normal data.

   * Small Sample Sizes: They can be applied to small sample sizes where parametric tests may not be appropriate.

   * Versatility: Non-parametric tests can be used for both continuous and categorical data.

* Common Non-Parametric Tests
   * Mann-Whitney U Test: Used to compare the medians of two independent groups.
   * Wilcoxon Signed-Rank Test: Used to compare the medians of two paired groups.
   * Kruskal-Wallis Test: Used to compare the medians of three or more independent groups.
   * Friedman Test: Used to compare the medians of three or more paired groups.
   * Spearman Rank Correlation Coefficient: Used to measure the correlation between two ranked variables.
   * Chi-Square Test: Used to test for independence between two categorical variables.

* Advantages of Non-Parametric Tests
  * Flexibility: Non-parametric tests can be applied to a wide variety of data types and are not limited by the assumptions of parametric tests.

  * Simple to Use: They often involve ranking data, which is simpler than dealing with raw data and complex calculations.

  * Robustness: Non-parametric tests are less affected by outliers and skewed distributions, making them robust in practical applications.

  * Versatile with Small Samples: They can be effectively used with small sample sizes where parametric tests may not provide reliable results.

* Disadvantages of Non-Parametric Tests
   * Less Power: Non-parametric tests are generally less powerful than parametric tests, meaning they may require larger sample sizes to detect a true effect.

   * Less Information: Because non-parametric tests focus on ranks rather than actual values, they may lose some information about the data's distribution.

   * Difficult Interpretation: The results of non-parametric tests, such as rank sums, may be less intuitive to interpret than means or variances from parametric tests.

This approach is particularly useful when the theoretical distribution of the statistic is complex or unknown, or when sample sizes are too small to rely on traditional parametric methods.

**How Bootstrap Methods Work**

Bootstrap methods involve the following steps:

   * Resampling: Generate a large number of bootstrap samples from the original dataset. Each bootstrap sample is created by randomly sampling with replacement from the original data, so some observations may be repeated in a bootstrap sample, while others may be omitted.

   * Compute Statistic: For each bootstrap sample, calculate the statistic of interest (e.g., mean, median, variance, regression coefficient).

   * Bootstrap Distribution: The collection of calculated statistics from all bootstrap samples forms the bootstrap distribution, which approximates the sampling distribution of the statistic.

   * Inference: Use the bootstrap distribution to estimate standard errors, confidence intervals, and p-values.

* Why Use Bootstrap Methods?
   * No Assumptions About Distribution: Bootstrap methods do not rely on the assumption that the data are normally distributed or follow any specific distribution.

   * Flexibility: They can be applied to a wide range of statistics beyond the mean, including medians, variances, and regression coefficients.

   * Useful for Small Samples: Bootstrap methods are particularly valuable when dealing with small sample sizes where traditional inferential methods might not be reliable.

   * Confidence Intervals: Bootstrap can be used to generate confidence intervals for statistics, providing a way to quantify the uncertainty of estimates.

* Types of Bootstrapping:
   * Non-parametric Bootstrap: Does not rely on any assumptions about the underlying distribution of the data.
   * Parametric Bootstrap: Assumes a specific parametric form for the underlying distribution.
   * Bootstrap with Smoothing: Uses techniques like kernel density estimation to smooth the bootstrap distribution.

* Advantages of Bootstrapping:
   * Versatility: Can be applied to a wide range of statistical problems.
   * Robustness: Less sensitive to assumptions about the underlying distribution.
   * Simplicity: Relatively easy to implement, even with complex data structures.
   * Non-Parametric Nature: Can be used with non-parametric statistics.
* Disadvantages of Bootstrapping:
   * Computational Intensity: Can be computationally intensive, especially for large datasets.
   * Sensitivity to Sample Size: May not perform well with small sample sizes.
   * Bias: Can introduce bias if the original sample is biased.

* Applications of Bootstrap Methods
   * Confidence Intervals: Bootstrap methods are often used to construct confidence intervals for population parameters when the sample size is small or when the distribution is unknown.

      * Example: Constructing a 95% confidence interval for the median of a dataset.
   * Hypothesis Testing: Bootstrap methods can be used to perform hypothesis tests by comparing the bootstrap distribution of the test statistic under the null hypothesis to the observed value.

      * Example: Testing the equality of two population means using the bootstrap distribution of the difference in means.
   * Model Validation: Bootstrap methods are used to estimate the accuracy of predictive models by resampling the training data and evaluating model performance on the bootstrap samples.

      * Example: Using bootstrap to estimate the standard error of a regression coefficient.
   * Bias Estimation: Bootstrap can be used to estimate and correct the bias of an estimator by comparing the mean of the bootstrap distribution to the observed statistic.

      * Example: Estimating the bias of the sample mean in a small dataset.
