### Introduction to Descriptive Statistics

* Descriptive statistics is a branch of statistics focused on summarizing and describing the main features of a data set. Unlike inferential statistics, which aims to make predictions or inferences about a population based on a sample, descriptive statistics simply provides a clear and concise overview of the data.

image.png

**Why it is used?**
* Descriptive statistics are used to simplify and summarize large data sets, providing a clear overview of key features. They help in:

1. Simplifying Data: Breaking down complex data into understandable metrics like mean, median, and standard deviation.
2. Summarizing Data: Offering quick insights into central tendencies, variability, and overall distribution.
3. Comparing Data: Facilitating comparisons between different data sets or periods.
4. Identifying Trends: Revealing patterns, trends, and anomalies.
5. Data Cleaning: Spotting inconsistencies and errors.
6. Exploring Data: Assisting in the initial analysis of data before applying more complex techniques.
7. Decision Making: Informing decisions with summarized metrics.

#### Mode

* The mode is the value that appears most frequently in the data set. A data set may have one mode, more than one mode, or no mode if no value repeats.
* Formula:
   * Mode is the value with the highest frequency.
* Example:

```python
from collections import Counter

def find_mode(data):
  """Finds the mode of a list of numbers.

  Args:
    data: A list of numbers.

  Returns:
    The mode of the data, or a list of modes if there are multiple.
  """

  # Count the occurrences of each number
  counts = Counter(data)

  # Find the maximum count
  max_count = max(counts.values())

  # Return the numbers with the maximum count
  return [num for num, count in counts.items() if count == max_count]

# Example usage
numbers = [1, 2, 3, 2, 4, 2, 5]
mode = find_mode(numbers)
print("Mode:", mode)
```

### Measures of Dispersion

* Measures of dispersion describe the spread or variability within a data set. They provide insights into how much the data values differ from the central value (e.g., mean or median). Key measures of dispersion include range, variance, standard deviation, and interquartile range (IQR).

#### Range

* The range is the difference between the maximum and minimum values in a data set. It provides a simple measure of spread but can be sensitive to outliers.

image.png

* Interpretation of Range
    * Large Range: A large range indicates that the data points are widely spread out. This suggests a greater degree of variability or uncertainty in the data.
    * Small Range: A small range means that the data points are clustered closely together. This indicates a smaller degree of variability or uncertainty.

* Uses of Range
    * Quick measure of spread: The range is easy to calculate and provides a quick estimate of the spread of the data.
    * Identifying outliers: The range can be used to identify potential outliers, which are data points that are significantly different from the rest of the data.    
* Limitations of Range
    * Sensitivity to outliers: The range is highly sensitive to outliers, as they can significantly affect the maximum or minimum values.
    * Limited information: The range only provides information about the extreme values, not the distribution of the data in between.

* Formula:

     * Range

* Example:

```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50]

# Calculate range
data_range = np.max(data) - np.min(data)
print(f"Range: {data_range}")
```

#### Variance

* Variance is a statistical measure that quantifies how much a set of data points deviate from their mean. In simpler terms, it measures how spread out the data is. A high variance indicates that the data points are widely dispersed, while a low variance means they are clustered closely together.
* Interpretation of Variance
    * High Variance: A high variance means that the data points are widely spread out from the mean. This can indicate a large degree of variability or uncertainty in the data.
    * Low Variance: A low variance means that the data points are clustered closely together around the mean. This suggests a smaller degree of variability or uncertainty.

* Uses of Variance

    * Hypothesis testing: To determine if differences between sample means are statistically significant.
    * Regression analysis: To measure the variability of the residuals, which are the errors between the predicted and actual values.
    * Quality control: To monitor process variation and identify when a process is out of control.
    * Investment analysis: To assess the risk associated with an investment by measuring the volatility of returns.

#### Standard Deviation

* The standard deviation is the square root of the variance. It provides a measure of spread in the same units as the data and is more interpretable than variance.

image.png

* Interpretation of Standard Deviation
    * High Standard Deviation: A high standard deviation indicates that the data points are widely spread out from the mean. This suggests a large degree of variability or uncertainty in the data.
    * Low Standard Deviation: A low standard deviation means that the data points are clustered closely together around the mean. This indicates a smaller degree of variability or uncertainty.
* Uses of Standard Deviation
Standard deviation is used in various statistical applications, including:

    * Hypothesis testing: To determine if differences between sample means are statistically significant.
    * Regression analysis: To measure the variability of the residuals, which are the errors between the predicted and actual values.
    * Quality control: To monitor process variation and identify when a process is out of control.
    * Investment analysis: To assess the risk associated with an investment by measuring the volatility of returns.
    * Normal distribution: In a normal distribution, approximately 68% of the data points fall within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.

* Formula:

image.png

 * Example:

```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50]

# Calculate standard deviation
std_dev = np.std(data)
print(f"Standard Deviation: {std_dev}")
```

#### Interquartile Range (IQR)

* The interquartile range measures the spread of the middle 50% of the data. It is the difference between the third quartile (Q3) and the first quartile (Q1). It is less sensitive to outliers than the range.

image.png

* Interpretation of IQR
    * Large IQR: A large IQR indicates that the data points are widely spread out in the middle of the dataset. This suggests a greater degree of variability or uncertainty in the middle 50% of the data.
    * Small IQR: A small IQR means that the data points in the middle of the dataset are clustered closely together. This suggests a smaller degree of variability or uncertainty in the middle 50% of the data.

* Uses of IQR
    * Robust measure of spread: IQR is less sensitive to outliers than the range, making it a more robust measure of dispersion.
    * Box plots: IQR is used in box plots to represent the middle 50% of the data.
    * Outlier detection: IQR can be used to identify outliers in a dataset. Outliers are data points that are significantly different from the rest of the data.

* Formula:
    * Order the data: Arrange the data points in ascending order.
    * Identify quartiles:
       * First quartile (Q1): The median of the lower half of the data.
       * Third quartile (Q3): The median of the upper half of the data.
    * Calculate IQR: Subtract Q1 from Q3:
       * IQR = Q3  Q1
* Example:

```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50]

# Calculate IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print(f"Interquartile Range (IQR): {IQR}")
```

### Measures of Relative Standing

#### Percentiles

* Percentiles indicate the relative standing of a value within a data set. The
p-th percentile is the value below which
p% of the data fall. For example, the 25th percentile (Q1) is the value below which 25% of the data lie.

image.png

* Common Percentiles
    * Median: The 50th percentile, representing the middle value of the dataset.
    * Quartiles: The 25th, 50th, and 75th percentiles, dividing the data into four equal parts.
    * Deciles: The 10th, 20th, 30th, ..., 90th percentiles, dividing the data into ten equal parts.
* Uses of Percentiles
    * Comparing data: Percentiles can be used to compare different datasets or different values within the same dataset.
    * Identifying outliers: Percentiles can help identify outliers by looking at extreme percentiles (e.g., the 1st and 99th percentiles).
    * Measuring dispersion: Percentiles can be used to measure the spread of data, especially when dealing with skewed distributions.
    * Understanding distributions: Percentiles can provide a better understanding of the shape of a distribution, including its skewness and kurtosis.

* Formula: To find the
p-th percentile, you need to sort the data and locate the value corresponding to the percentile rank.
* Example:

```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50, 60, 70]

# Calculate percentiles
percentile_25 = np.percentile(data, 25)
percentile_50 = np.percentile(data, 50)  # Median
percentile_75 = np.percentile(data, 75)

print(f"25th Percentile: {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile: {percentile_75}")
```

#### Quartiles

* There are three primary quartiles:

    * First quartile (Q1): The value that separates the lower 25% of the data from the upper 75%.
    * Second quartile (Q2): Also known as the median, it separates the lower 50% of the data from the upper 50%.
    * Third quartile (Q3): The value that separates the lower 75% of the data from the upper 25%.

* To calculate quartiles, the data must first be sorted in ascending order. Then, the quartiles can be found using the following methods:

    * For a dataset with an odd number of data points:
       * The median (Q2) is the middle value.
       * Q1 is the median of the lower half of the data, excluding the median.
       * Q3 is the median of the upper half of the data, excluding the median.
    * For a dataset with an even number of data points:
       * The median (Q2) is the average of the two middle values.
       * Q1 is the median of the lower half of the data.
       * Q3 is the median of the upper half of the data.   

* Uses of Quartiles
    * Understanding data distribution: Quartiles provide insights into the shape and spread of a dataset.
    * Identifying outliers: Outliers can be identified by examining the distance between the whiskers in a box plot, which are based on the quartiles.
    * Comparing datasets: Quartiles can be used to compare different datasets or different groups within a dataset.
    * Calculating the interquartile range (IQR): The IQR is calculated as Q3 - Q1 and provides a measure of the spread of the middle 50% of the data.

* Example:

```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50, 60, 70]

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)

print(f"First Quartile (Q1): {Q1}")
print(f"Second Quartile (Q2, Median): {Q2}")
print(f"Third Quartile (Q3): {Q3}")
```

#### Z-Scores

* The z-score measures how many standard deviations a data point is from the mean. It provides a way to compare data points from different distributions.

image.png

* Interpretation of Z-Scores
     * Positive z-score: Indicates that the data point is above the mean.
     * Negative z-score: Indicates that the data point is below the mean.
     * Magnitude of the z-score: The magnitude of the z-score indicates how many standard deviations the data point is from the mean. For example, a z-score of 2 means the data point is 2 standard deviations above the mean.

* Uses of Z-Scores
    * Comparing data: Z-scores allow for comparisons between data points from different datasets or variables, as they are standardized to a common scale.
    * Identifying outliers: Outliers can be identified by looking at z-scores that are significantly larger or smaller than the rest of the data.
    * Normal distribution: Z-scores are used to calculate probabilities and percentiles in a normal distribution.
    * Hypothesis testing: Z-scores are used in various hypothesis testing procedures, such as the z-test.   

* Formula:

image.png

where
     * x is the individual data point
     *  is the mean of the dataset
     *  is the standard deviation of the dataset

* Example:


```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50, 60, 70]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate z-scores
z_scores = [(x - mean) / std_dev for x in data]
print(f"Z-Scores: {z_scores}")
```

**Summary**
* Percentiles: Show the relative position of a value within the data set. Useful for understanding the distribution of data.
* Quartiles: Divide data into four equal parts. Help in understanding the spread and central tendencies.
* Z-Scores: Indicate how many standard deviations a data point is from the mean. Useful for comparing values across different distributions.
* These measures provide a way to assess individual data points relative to the overall distribution, making it easier to understand their position and significance within the data set.

### Measures of Shape

* Measures of shape describe the distribution of data points in a data set, focusing on aspects such as symmetry and the heaviness of the tails. Key measures of shape include skewness and kurtosis.

#### Skewness

* Skewness measures the asymmetry of the data distribution. It indicates whether the data distribution leans to the left or right.

image.png

* Types of Skewness
    * Positive Skewness (Right-Skewed): The tail on the right side of the distribution is longer. This means that there are a few extreme values on the high end of the data.
    * Negative Skewness (Left-Skewed): The tail on the left side of the distribution is longer. This means that there are a few extreme values on the low end of the data.
    * Symmetrical Skewness: The distribution is symmetrical, with both tails being of equal length.

* Uses of Skewness
    * Understanding data distribution: Skewness provides insights into the shape of a distribution and how the data is distributed.
    * Identifying outliers: Skewness can help identify outliers, as they often contribute to the asymmetry of the distribution.
    * Choosing statistical tests: The choice of statistical tests may be influenced by the skewness of the data. For example, the t-test is more robust to skewness than the z-test.

#### Kurtosis

* Kurtosis measures the "tailedness" or the peakedness of the data distribution. It indicates whether the data has heavy tails or light tails compared to a normal distribution.

image.png

* Types of Kurtosis
    * Leptokurtic: A distribution with heavy tails, meaning there are more extreme values than in a normal distribution. It is also known as a "peaked" distribution.
    * Mesokurtic: A distribution with tails similar to a normal distribution.
    * Platykurtic: A distribution with light tails, meaning there are fewer extreme values than in a normal distribution. It is also known as a "flattened" distribution.

* Interpretation of Kurtosis
    * Positive excess kurtosis (leptokurtic): The distribution has heavy tails with more extreme values than a normal distribution.
    * Negative excess kurtosis (platykurtic): The distribution has light tails with fewer extreme values than a normal distribution.
    * Zero excess kurtosis (mesokurtic): The distribution has tails similar to a normal distribution.

* Uses of Kurtosis
    * Understanding data distribution: Kurtosis provides insights into the shape of a distribution and how the data is distributed.
    * Identifying outliers: Kurtosis can help identify outliers, as they often contribute to heavy tails.
    * Choosing statistical tests: The choice of statistical tests may be influenced by the kurtosis of the data. For example, some tests are more robust to kurtosis than others.

**Summary**

* Kurtosis: Measures the heaviness of the tails of the distribution. Positive kurtosis indicates heavy tails and a sharp peak, while negative kurtosis indicates light tails and a flatter peak.
* These measures help in understanding the shape and distribution characteristics of the data, which can be crucial for selecting appropriate statistical methods and interpreting results accurately.

### Application in Real-World Scenarios

* Descriptive statistics are widely used in various real-world scenarios to summarize, analyze, and make decisions based on data. Here are some practical applications with coding examples:

1. Healthcare
* In healthcare, descriptive statistics are used to summarize patient data, treatment outcomes, and health metrics.

Example: Analyzing Patient Blood Pressure

Let's analyze blood pressure readings of patients to understand their distribution.

```python
import numpy as np
import pandas as pd

# Sample blood pressure readings (systolic, diastolic)
data = {
    'Systolic': [120, 130, 125, 140, 135, 145, 150, 140, 135, 130],
    'Diastolic': [80, 85, 82, 90, 88, 92, 95, 88, 86, 84]
}
df = pd.DataFrame(data)

# Descriptive statistics
mean_systolic = df['Systolic'].mean()
mean_diastolic = df['Diastolic'].mean()
median_systolic = df['Systolic'].median()
median_diastolic = df['Diastolic'].median()
std_dev_systolic = df['Systolic'].std()
std_dev_diastolic = df['Diastolic'].std()

print(f"Mean Systolic: {mean_systolic}")
print(f"Mean Diastolic: {mean_diastolic}")
print(f"Median Systolic: {median_systolic}")
print(f"Median Diastolic: {median_diastolic}")
print(f"Standard Deviation Systolic: {std_dev_systolic}")
print(f"Standard Deviation Diastolic: {std_dev_diastolic}")
```

2. Education
* In education, descriptive statistics help analyze student performance, test scores, and attendance.

Example: Analyzing Student Test Scores

Let's analyze the scores of students in a recent test.

```python
import numpy as np

# Sample test scores
scores = [75, 85, 90, 78, 92, 88, 76, 80, 84, 95]

# Descriptive statistics
mean_score = np.mean(scores)
median_score = np.median(scores)
std_dev_score = np.std(scores)
range_score = np.max(scores) - np.min(scores)

print(f"Mean Score: {mean_score}")
print(f"Median Score: {median_score}")
print(f"Standard Deviation of Scores: {std_dev_score}")
print(f"Range of Scores: {range_score}")
```

3. Business Analytics
* In business, descriptive statistics help analyze sales data, customer behavior, and market trends.

Example: Analyzing Sales Performance

Let's say you have sales data for a retail store and want to analyze the performance over a month.

```python
import pandas as pd

# Sample sales data for a month
data = {
    'Sales': [1200, 1500, 1800, 2000, 2500, 3000, 2800, 2200, 2300, 2700, 3200, 2900]
}
df = pd.DataFrame(data)

# Descriptive statistics
mean_sales = df['Sales'].mean()
median_sales = df['Sales'].median()
std_dev_sales = df['Sales'].std()
data_range = df['Sales'].max() - df['Sales'].min()

print(f"Mean Sales: {mean_sales}")
print(f"Median Sales: {median_sales}")
print(f"Standard Deviation of Sales: {std_dev_sales}")
print(f"Range of Sales: {data_range}")
```
