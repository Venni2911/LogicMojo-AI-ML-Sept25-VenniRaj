```python
Descriptive statistics
Mean - avg
Median - middle value in ordered set
Mode - value with highest frequency
Range = diff between max value - min value
variance - measure that quantifies how much a set of data points deviate from their mean - how data spread out
std deviation = square root of variance -
high std deviation means data is spread out - more uncertainity - data points away from mean
low std deviation means - less spread out - data points close to mean
IQ range = 75th percentile - 25th percentile
Z score - how many std deviations away from mean
+ve z score = above mean
-ve z score = below mean
Skew - show the outliners .if the data is skewed that means it has outliners in certain direction
Kurtosis - tells how your tail looks
normal distribution = median, mode, mean are same
```
