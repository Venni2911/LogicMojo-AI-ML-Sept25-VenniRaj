M1 produces 60% of items, M2 produces 40%.

Defective rates: 3% for M1, 5% for M2.
If an item is found defective, what is the probability it was produced by M1?




```python
P (M1/D)  = P (D/M1) * p (M1) /( P(D/M1) * P(M1) + P (D/M2) * P (M2))
= 0.03 * 0.6 / (0.03 * 0.6 + 0.05 * 0.4 ) = 0.018/ 0.038 = 0.47
```

Specificity (true negative rate) = 90%

```python
P (D/+) = P(+/D) * P (D) / P(+/D) * P (D) + p(+/~D) * P(~D)

= 0.95 * 0.01 / 0.95 * 0.01 + 0.10 * 0.99
= 0.087
## +, - , D/+ ,D/-
```

```python
2. Binomial distribution -> Estimate the successes in n independent bernoulli trials

Eg -> Estimate the p of heads in 10 tossed of coin

Mean -> E(X) = np
variance -> var(X) = np(1-p)
```

```python
3. Poisson Distribution -> models th events that occur in a specific interval of time/ space

eg -> number of customers that visit a shop in 1 hour


Mean -> E(X) -> Lambda - average visits in an hour

Var(X) = Lambda
```

```python
4. Geometric Distribution -> number of trials for 1st success

eg -> no of times you toss a coin till you get a heads

Mean -> = 1/p
Var -> (1-p)/p
```

```python
5. Negative Binomial distribution -> no of trials to get r successes

eg -> no of coin tosses to get 5 heads

Mean = r/p
variance = r(1-p)/p^2
```
